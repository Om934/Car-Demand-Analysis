# Conduct in-depth statistical analysis of the trips data and give your observations

```python
pd.unique(df['Month'])
```

```python
d = {}
for i in pd.unique(df['Monthname']):
    count = 0
    for j in df["Monthname"]:
        if i == j:
            count = count + 1
        d[i] = count
```

```python
import matplotlib.pyplot as plt

plt.bar(d.keys() , d.values() , width = 0.2 , color = "cyan")

plt.xlabel("Month_name")

plt.ylabel("No.of Trips")

plt.show()
```

```python
# based on the obsevation in new york city April month has less number of trips and september moth has high number of trips
```
